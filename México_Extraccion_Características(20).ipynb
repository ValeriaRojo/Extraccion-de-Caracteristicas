```python
#Cargamos librerías (checar versión de python en lo local)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from funpymodeling.exploratory import freq_tbl
```

```python
#Carga desde un archivo xlsx sin índice
Mex_Limpio = pd.read_csv('Mexico_Datos_Limpios_2.csv')
```

```python
#Obtenemos un análisis univariado de las variables categóricas
freq_tbl(Mex_Limpio)
```

```python
#Obtengo un análisis univariado de una variable "bathrooms_text"
table5=freq_tbl(Mex_Limpio['bathrooms_text'])
table5
```

```python
table5= table5.drop(['percentage', 'cumulative_perc'], axis=1)
table5
```

```python
#Obtenemos un filtro de los valores mas relevantes de las variables categórica
#Por ejemplo, en este caso eliminamos todos los 1 ue existen en la frecuencia
Filtro= table5[table5['frequency']>1]
Filtro
```

```python
#Convertimos la categoría en el índice para que la podamos graficar
Filtro_index= Filtro.set_index('bathrooms_text')
Filtro_index
```

```python
#Realizamos gráfico de battas de dataframe filtrado
Filtro_index.plot(kind = 'bar', width= 0.8, figsize=(10,4), color= "purple")
plt.title('Cantidad de baños')
plt.xlabel('Baños cantidad')
plt.ylabel('Frecuencia')
```
